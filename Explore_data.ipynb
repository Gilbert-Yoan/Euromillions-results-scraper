```python
import matplotlib.pyplot as plt
import numpy as np

#stacked histogram
plt.figure(figsize=(12, 6))
#Found here but starts at 0, since we start at 1, we will do +2
#https://stackoverflow.com/questions/27083051/matplotlib-xticks-not-lining-up-with-histogram
bins = max(max(inner_array) for inner_array in day_numbers) + 2
plt.hist(day_numbers, bins=np.arange(bins)-0.5, label=['Tuesday', 'Friday'], stacked=True, width=0.8, align='mid', color=["navy","blue"])
plt.xticks(range(bins),rotation=90)
#minus 1 since we start at 0
plt.xlim(0,bins-1)
plt.title('Frequency by Numbers and Day of the week')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
